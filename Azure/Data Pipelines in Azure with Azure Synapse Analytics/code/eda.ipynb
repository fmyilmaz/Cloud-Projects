```python
# spliting the name into first name and last name

def split_name(name: str) -> tuple:
    """ Split the name into first name and last name 
    Args:
        name (str): Name of the athlete
        Returns:
        tuple: (first_name, last_name)
        
    Examples:
    >>> split_name('Michael Phelps')
    ('Michael', 'Phelps')
    """
    parts = name.split(sep=' ', maxsplit=1)
    if len(parts) == 2:
        last_name, first_name = parts
        if last_name.isupper():
            return first_name, last_name  
    return name, ''
```

```python
athletes[['FirstName', 'LastName']] = athletes['Name'].apply(lambda x: pd.Series(split_name(x)))
```
